# Breast Cancer Prediction Using Logistic Regression
**Dataset Source**: YBI Foundation  
**Tool**: Python, Scikit-learn, Pandas  
**Model Used**: Logistic Regression

## Introduction
The goal of this project is to build a machine learning model to predict whether a tumor is **malignant (M)** or **benign (B)** using various cell features from digitized images.  
Accurate prediction can assist in early diagnosis and potentially save lives.

We'll use a **Logistic Regression** classifier for this binary classification problem.

```python
import pandas as pd
```

```python
cancer = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv')
```

```python
cancer.head()
```

```python
cancer.info()
```

```python
cancer.describe()
```

## Dataset Description
- **Rows:** 569 samples
- **Target Variable:** `diagnosis` (M = Malignant, B = Benign)
- **Features:** 30 numerical features extracted from digitized images of fine needle aspirate (FNA) of breast mass.

Features include:
- Mean, standard error, and worst (largest) values for radius, texture, perimeter, area, smoothness, etc.
- Final column `Unnamed: 32` is completely empty and will be dropped.


```python
cancer.columns
```

```python
y = cancer['diagnosis']
```

```python
x = cancer.drop(['id','diagnosis','Unnamed: 32'], axis = 1)
```

## Data Cleaning
We drop the following:
- `id`: Not useful for prediction  
- `Unnamed: 32`: Empty column  
- `diagnosis`: Separated as the target variable `y`


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.7, random_state=2529)
```

## Train-Test Split
We'll split 70% of the data for training and 30% for testing.


```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=5000)
```

```python
model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

cr = classification_report(y_test, y_pred)
print("Classification Report:\n", cr)
```

```python
# Visualizing the confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', xticklabels=['B','M'], yticklabels=['B','M'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

## Conclusion
The Logistic Regression model was successful in predicting breast cancer diagnosis with an accuracy of **~96%**.  

The model’s ability to correctly classify both malignant and benign tumors makes it a reliable tool for medical prediction tasks. However, future improvements could include testing other classification algorithms such as Random Forest or Support Vector Machines for better robustness.

Overall, this model demonstrates how machine learning can support early detection in healthcare applications.
